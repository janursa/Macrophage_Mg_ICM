```python
import os
from tools import dirs
from tools.tools import calibrate, edit_matlab_model
import json
%load_ext autoreload
%autoreload 2
%config Completer.use_jedi = False
showOriginalModelString = True
```

```python
target_package = 'P3'
###-------- import --------###
from models.params import free_params_p, fixed_params
from data.observations import observations
from models.models import Macrophage
### ----------------------###
edit_matlab_model(input_file=dirs.dir_matlab_model,output_file=dirs.dir_model)
free_params = free_params_p[target_package]
print(free_params)
model_obj = Macrophage(observations)

def callback(xk, convergence):
    params = {**fixed_params}
    keys = list(free_params.keys())
    for ii in range(len(keys)):
        params[keys[ii]] = xk[ii]
    error = model_obj.run(params = params)
    if  error < 0.01:
        return True
class Strategies:
    best1bin = 'best1bin'
    rand1exp = 'rand1exp'
inferred_params = calibrate(model = model_obj,fixed_params = fixed_params, free_params=free_params, n_proc=1,disp=True,max_iters=200,strategy=Strategies.best1bin,callback=callback)
with open(dirs.dir_calib_output,'w') as f:
    f.write(json.dumps(inferred_params))
os.system('say "Hey Matin, calibration is done, come back"')
```
